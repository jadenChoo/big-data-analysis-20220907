# Tokenize

```python
import nltk
import string
```

```python
from nltk.corpus import stopwords
```

```python
#nltk.download('stopwords')
```

```python
def tokenize(text):
    text = text.lower()
    stemmer = nltk.stem.SnowballStemmer('english')
    
    for token in nltk.word_tokenize(text):
        if token in string.punctuation:
            continue
        elif token in stopwords.words('english'):
            continue
            
        yield stemmer.stem(token)
```

```python
for doc in corpus[:2]:
    
    token = list(tokenize(doc))
    
    print(doc)
    print(token)
    print()
```

## Clustering

```python
from gensim.matutils import sparse2full
```

```python
vectors = []
```

```python
for tokens in tokenized_corpus:
    vec = lexicon.doc2bow(tokens)
    vec = [(x[0], 1) for x in vec]
    vec = tfidf[vec]
    
    vec = sparse2full(vec, len(lexicon))
    
    vectors.append(vec)
```

```python
distance = nltk.cluster.cosine_distance
```

```python
kmeans = nltk.cluster.KMeansClusterer(
    num_means = 6,
    distance = distance,
    avoid_empty_clusters = True,
)
```

```python
def extract_main_keywords(group, n=10):
    entries = classified[group]
    vec_list = [x[2] for x in entries]
    
    matrix = np.array(vec_list)
    score_sum = np.sum(matrix, axis=0)
    
    paired = [(x, i) for i, x in enumerate(score_sum)]
    paired.sort(reverse=True)
    
    top = paired[:n]
    top = [x[1] for x in top ]
    
    keywords = [lexicon[x] for x in top]
    
    return keywords
```

```python
main_keywords = defaultdict(list)
```

```python
for group in range(len(classified)):
    main_keywords[group] = extract_main_keywords(group)
    
    print(f'* Group {group}')
    print(main_keywords[group])
    print()
```
